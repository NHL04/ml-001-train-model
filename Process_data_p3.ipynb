```python
# Replace NaN values with 'khongco'
df.fillna('unassigned', inplace=True)

# Function to process each value
def process_value(value):
    names = value.split('&')
    if len(names) == 1:
        return names[0]
    elif len(set(names)) == 1:
        return names[0]
    else:
        return 'multi'
# Apply the function to the DataFrame
df['GVLT'] = df['GVLT'].apply(process_value)
df['GVTH'] = df['GVTH'].apply(process_value)
```

```python
df['GVTH'] = df['GVTH'].replace('nan', 'unassigned')
df['GVLT'] = df['GVLT'].replace('nan', 'unassigned')
```
